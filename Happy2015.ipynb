```python
import numpy as np
import matplotlib.pyplot as plt

# Beispiel-Daten
x = np.array([1, 2, 3, 4, 5])
y = np.array([1, 4, 9, 16, 25])

# Polynomiale Regression
coefficients = np.polyfit(x, y, 2)  # Polynom 2. Grades
p = np.poly1d(coefficients)

# Daten zur Darstellung generieren
x_fit = np.linspace(0, 6, 100)
y_fit = p(x_fit)

# Daten und Regression plotten
plt.scatter(x, y, label="Daten")
plt.plot(x_fit, y_fit, label="Polynomiale Regression", color='red')
plt.legend()
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

```python
import numpy as np
import pandas as pd
import statsmodels.formula.api as smf
import matplotlib.pyplot as plt

# Beispiel-Daten
data = pd.DataFrame({'x': np.array([1, 2, 3, 4, 5]),
                     'y': np.array([1.2, 3.5, 8.1, 15.2, 24.8])})

# Polynomiale Regression mit einer Formel (Polynom 2. Grades)
model = smf.ols(formula="y ~ x + I(x**2)", data=data).fit()

# Daten zur Darstellung generieren
x_fit = np.linspace(0, 6, 100)
y_fit = model.predict(pd.DataFrame({'x': x_fit, 'I(x**2)': x_fit**2}))

# Daten und Regression plotten
plt.scatter(data['x'], data['y'], label="Daten")
plt.plot(x_fit, y_fit, label="Polynomiale Regression (2. Grades)", color='red')
plt.legend()
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```
