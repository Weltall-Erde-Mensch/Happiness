```python
import pandas as pd
import numpy as np
import statsmodels.formula.api as smf
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Data/2015.csv', sep=',')
df.head()
```

```python
df.isnull().sum()
```

```python
df_plot = df[['Happiness Score',
              'Economy (GDP per Capita)',
              'Family',
              'Health (Life Expectancy)',
              'Freedom',
              'Trust (Government Corruption)',
              'Generosity']]

df_plot = df_plot.rename(columns={'Happiness Score':'Happiness_score', "Economy '('GDP per capita')'":'GDP_per_capita','Family':'Family',
                                  'Health (Life Expectancy)':'Life_expectancy','Freedom':'Freedom','Trust (Government Corruption)':'Government_corruption',
                                  'Generosity':'Generosity'})

pd.plotting.scatter_matrix(df_plot, figsize=[20,15]);
```

```python
model = smf.ols(formula="Family ~ 'Economy (GDP per capita)'", data=df_plot)
results = model.fit()
results.summary()
```
